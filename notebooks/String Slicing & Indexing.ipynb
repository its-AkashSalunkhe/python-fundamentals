# String Slicing

```python
name = "Ram gopal venu gopal ayyar"
```

```python
name[4:23]
```

```python
name[20:32]
```

```python
name[3:25:2]
```

# String Indexing

```python
name[:]    #prints string as it is
```

```python
name[::2]  # This is called forword Indexing
```

```python
name[::-2]  # This is called Backword Indexing
```

```python
name[::-1]  # This Reverses the String
```
