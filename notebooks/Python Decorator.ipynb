# Python Decorator

```python
def div(a,b):
    print(a/b)


div(4,2)
```

```python
def div(a,b):
    print(a/b)


div(2,4)
```

```python
# This is important here, with the same fucntion if we change the input then it directly affect the output|
```

```python
# In python Decorator ===> we change(update) the code without affecting the output
```

```python
def div(a,b):
    if a<b:
        a,b = b,a
    print(a/b)

div(4,2)
```

```python
def div(a,b):
    if a<b:
        a,b = b,a
    print(a/b)

div(2,4)
```

```python
# See in the above code we change(update) the code without affecting the output
```

```python
def decorator(func):
    def wrapper():
        print("Something happen before")
        func()
        print("Something happen after")
    return wrapper


@decorator 
def say_hello():  
    print("Hello")

say_hello()
```
