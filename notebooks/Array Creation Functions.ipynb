# 2. Array Manipulation Functions

```python
    # Reshape an Array


a1 = np.array([1,2,3])
reshaped = np.reshape(a1, (3,1))             # this will reshape to 1 x 3
print("Reshaped Array :\n",reshaped)
```

```python
    # Flatten an Array


f1 = np.array([[1,2],[3,4]])
flattened = np.ravel(f1)                      # this will flatten to 1D Array
print("Flattened Array : ",flattened)
```

```python
   #  Transpose of an Array


e1 = np.array([[1,2],[3,4]])
print(np.transpose(e1))
```

```python
   # Stack Arrays Vertically


a2 = np.array([1,2,3])
b2 = np.array([4,5,6])
print("Stack Array:\n ", np.vstack([a2,b2]))
```

```python
  # Stack Arrays horizontally


a2 = np.array([1,2,3])
b2 = np.array([4,5,6])
print("Stack Array:\n ", np.hstack([a2,b2]))
```
