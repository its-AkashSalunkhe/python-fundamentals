# Set Operations

# 1. Union

```python
a1 = {1,2,3,4,5,6}
b1 = {5,6,7,8}
c1 = {7,8,9,0}
```

```python
a1.union(b1)   # Duplicate values are not allowed in sets
```

```python
a1.union(b1,c1)
```

```python
# we can also use '|' sign to denote 'Union"

a1 | b1 | c1
```

# 2. Intersection

```python
a1.intersection(b1)
```

```python
b1.intersection(c1)
```

```python
# we can also use '&' sign to denote 'intersection"

a1 & b1
```

# 3. Differemce

```python
a1.difference(b1)
```

```python
b1.difference(a1)       # Expect common elemenet all the other elements will print from one set only
```

```python
# we can also use '-' sign to denote 'Difference"


a1 - b1
```

```python
b1 - c1
```

```python
a1.symmetric_difference(b1)    # Expect common elemenet all the other elements will print from both sets
```

```python
# we can also use '^' sign to denote 'symmetric_difference"

a1 ^ b1
```

```python
print(a1)
print(b1)
print(c1)
```

```python
a1 - b1
```

```python
a1    
```

```python
a1.difference_update(b1) 
```

```python
print(a1)
print(b1)
print(c1)
```
