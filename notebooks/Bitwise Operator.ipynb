```python
# Just like OR Truth table. Only 0 - 0 = 0 rest all are 1
```

# Bitwise XOR ( ^ )

```python
# Just like XOR Truth table.  0 - 0 = 0 and 1 - 1 = 0 rest all are 1
```

```python
12 ^ 13
```

```python
# because 1100 and 1101 = 0001, which is 13 hence the asnwer is 1  ( left side 0 have no value and are not considered
```
