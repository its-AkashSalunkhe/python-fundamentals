# 'as' Keyword

```python
math.pow(3,2)      # pow = Power of
```

```python
# now if I used m.pow(3,2) it will give me an error because m is not define as keyword

m.pow(3,2)
```

```python
# But you can define the keyword as

import math as m
m.pow(3,3)                # Now you can use 'm.' everywhere instead of writing "math' everytime
```

```python
math.pow(3,5)        # even if you define 'm' as keyword still you can use "math." 
```

# 'from' Keyword

```python
import math as m

print(m.pow(3,2))
print(m.sqrt(25))
print(m.floor(3.14))
print(m.ceil(3.14))
```

```python
# now instaed of doing all the time this I can simply do -

from math import pow, floor, ceil, sqrt

print(pow(3,2))
print(sqrt(10))
print(floor(3.14))
print(ceil(3.14))
```

# how to access all functions in maths module

```python
# we can access all the functions in any module using '*' sign

from math import *

print(pow(3,2))
print(sqrt(10))
print(floor(3.14))
print(ceil(3.14))
print(factorial(5))
print(gcd(2,4,6))
```
